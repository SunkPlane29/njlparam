```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import scienceplots

plt.style.use(['science', 'notebook', 'grid'])
```

```python
# importando os dados de o arquivo csv 
# Dados referentes ao nested sampling 
# com T_pc = 156 MeV \pm 1.5 MeV 
data = pd.read_csv("njlparameters_chain.csv")
data
```

```python
# plotando as distribuiÃ§Ãµes dos parÃ¢metros 
# Correlacionadas dos parametros 
plt.figure(figsize=(10, 6))
plt.scatter(data['GL2'], data['Lamb'], alpha=0.5, color='blue')
plt.xlabel('$G \Lambda^2$')
plt.ylabel('$\Lambda$ (GeV)')
plt.title('DistribuiÃ§Ã£o Correlacionada de G_S e Lambda')
plt.grid(True)
plt.show()
```
